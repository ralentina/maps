```python
# import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import data, set index & preview
pobData_df = pd.read_csv('data/pob_flot_y_perm.csv')
pobData_df.set_index('COMUNA', inplace=True, drop=True)
pobData_df.head(3)
```

```python
# get number of rows
pobData_df.shape[0]
```

```python
# create a single figure with 36 graphs in a six by six grid
fig, axes = plt.subplots(nrows=6, ncols=6, figsize=(32, 24))
# iterate through the rows of the data frame, plotting each raw as a separate line graph
# zip() creates a new data frame by pairing the values fed in the bracket, so, here, the index values of pobFlot and axes.flatten
# have not fully understood axes.flatten; I understand it collapses nested arrays into one dimension, so it helps to standardise the data
for row, ax in zip(pobData_df.index, axes.flatten()):
    sizes = [pobData_df.loc[row]['POB_FLOT_ESTIMADA_DIV12'],pobData_df.loc[row]['T_POB_U']]
    ax.pie(sizes, autopct='%1.1f%%', startangle=60, wedgeprops=dict(width=0.3))
    ax.set_title(row, fontsize=10)
plt.show()
```

```python
fig.savefig('pieCharts.eps', format='eps')
```
